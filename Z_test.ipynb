#Z-TEST

- Z-test is a statistical method to determine whether the distribution of the test statistics can be approximated by a normal distribution.
- It is the method to determine whether two sample means are approximately the same or different when their variance is known and the sample size is large (should be >= 30).

When to Use Z-test:

- The sample size should be greater than 30. Otherwise, we will use the t-test.

- Samples should be drawn at random from the population.

- The standard deviation of the population should be known.

- Samples that are drawn from the population should be independent of each other.

- The data should be normally distributed, however for large sample size, it is assumed to have a normal distribution.




##Level of significance:
 -  It means the degree of significance in which we accept or reject the null-hypothesis.

    Since in most of the experiments 100% accuracy is not possible for accepting or rejecting a hypothesis, so we, therefore, select a level of significance. It is denoted by alpha (∝).


##Steps to perform Z-test:

  - First, identify the null and alternate hypotheses.

 - Determine the level of significance (∝).
 - Find the critical value of z in the z-test using.
 - Calculate the z-test statistics. Below is the formula for calculating the z-test statistics.

          Z  =  (sample mean - population mean) / (standard deviation of the population/ sample size)
          
          
  -  Now compare with the hypothesis and decide whether to reject or not to reject the null hypothesis




##Types of Z test

  - Left -Tailed Test: In this test, our region of rejection is located to the extreme left of the distribution. Here our null hypothesis is that the claimed value is less than or equal to the mean population value. 

  - Right-tailed Test: In this test, our region of rejection is located to the extreme right of the distribution. Here our null hypothesis is that the claimed value is less than or equal to the mean population value.

  - Two-tailed test: In this test, our region of rejection is located to both extremes of the distribution. Here our null hypothesis is that the claimed value is equal to the mean population value.
 
- When a hypothesis is formed, the hypothesis testing is used to confirm how close the assumption is to reality. 


  


#When to use what?

A two-tailed test is appropriate if you want to determine if there is any difference between the groups you are comparing. For instance, if you want to see if Group A scored higher or lower than Group B, then you would want to use a two-tailed test. This is because a two-tailed test uses both the positive and negative tails of the distribution. In other words, it tests for the possibility of positive or negative differences.



A one-tailed test is appropriate if you only want to determine if there is a difference between groups in a specific direction. So, if you are only interested in determining if Group A scored higher than Group B, and you are completely uninterested in possibility of Group A scoring lower than Group B, then you may want to use a one-tailed test. The main advantage of using a one-tailed test is that it has more statistical power than a two-tailed test at the same significance (alpha) level. In other words, your results are more likely to be significant for a one-tailed test if there truly is a difference between the groups in the direction that you have predicted. This is because only one tail of the distribution is used for the test.

# One Tail Z-Test

Q1 - The population of all verbal GRE scores are known to have a standard deviation of 8.5. The
UW Psychology department hopes to receive applicants with a verbal GRE scores over 210.
This year, the mean verbal GRE scores for the 42 applicants was 212.79. Using a value of
α = 0.05 is this new mean significantly greater than the desired mean of 210?

###Us is the Sample mean
###Up is the Population mean


dataa distribution.png

Step 1 - identify the Null Hypothesis(H0) & Alternate Hypothesis(H1)

```python
# Standard deviation of the populaton is known. That is : 8.5
# Level of significance : 0.05(95%)
# no. of observation is > 30
# Sample mean is 212.79



''' 

Hypothesis for the above is as follows:

Null Hypothesis : The sample mean is significantly less or equal to 210.
             H0 = Us <= 210

Alternate Hypothesis : The sample mean is significantly greater than 210.
             H1 = Us > 210

'''
```

Step 2 - Determine the Level of Significance 

```python
'''

Level of significance given in the study problem is 0.05 that is with 95% of confidence we can say that the sample mean is greater or less than the sample mean.

'''
```

Step 3 - Determine the Critical Value



zones.png

```python
#Critical value actually divides the area under the curve into two zones. The two zones are as follows:

''''

 a. Acceptance Zone
 b. Reject Zone

 where the acceptance zone asks you to accept the null hypothesis & the reject zone asks you to reject the null hypothesis that is accept the alternate hypothesis.

The decision is made based on the z stat value. 
If the z stat value is more than the critical value it falls under Reject zone and if it is less than the critical value then it falls under acceptance zone in case of ISF.

'''

import numpy as np
import pandas as pd
from scipy.stats import norm
import math as m


sample_mean = 212.79
population_mean = 210
sample_size = 42
standard_deviation = 8.5

#Z-Statistic value
z_stat = ((sample_mean - population_mean) / (standard_deviation / m.sqrt(sample_size)))
print('z_statistic value is ',z_stat)


cv = norm.isf(0.05)
print('Critical value is',cv,' - (That is, if the Z-statistic value is more than 1.64 then we will reject the null hypothesis(H0) in case of ISF )','\n')


if z_stat > cv:
  print('Reject the null hypothesis as z statistic value is greater than the critical value thus, falls into the reject zone')
```

Conclusion - 

- Our observed value of z is 2.13 which is greater than the critical value of 1.64. We therefore
reject H0
.
- The sample mean is significantly greater than 210.


#Two-Tailed Z-Test

Q2  Suppose you start up a company that has developed a drug that is supposed to increase IQ.
You know that the standard deviation of IQ in the general population is 15. You test your
drug on 36 patients and obtain a mean IQ of 97.65. Using an alpha value of 0.05, is this IQ
significantly different than the population mean of 100?


- In this problem the agenda is to figure out that the sample mean is different from the population mean. Thus, we will go ahead with two tailed test as the IQ of the subjects is supposed to be increased due to the drug being tried on the subjects. However, the result could be anything that is the IQ might have increased or might not have had.

As it is directionless, two tailed Z-Test will be appropriate.

```python
#Standard deviation of the population = 15
# Population mean(Up) = 100
# sample size = 36
# sample mean(Us) = 97.65
# Level of Significance = 0.05 - 95%

'''

Hypothesis for the above is as follows:

Null Hypothesis - The sample mean IQ is equal to the population mean IQ
              (H0) = Us = Up

Alternate Hypothesis - The sample mean IQ is not equal to the population mean IQ
              (H1) = Us ≠ Up

'''
```

```python
sample_mean= 97.65
population_mean = 100
std_dev = 15
sample_size = 36


z_stat = ((sample_mean - population_mean) / (std_dev / m.sqrt(sample_size)))
print('z_statistic value is ',z_stat)


cv= norm.ppf(0.05/2)
print('critical value is', cv, '(1.96 is the critical value in the right tail and -1.96 on the left tail)','\n')


if z_stat > cv:
  print('Accept the null hypothesis(fail to reject the null hypothesis), as z stat falls into the acceptance zone. ')
```

Conclusion - 
- The rejection region contains values of z less than -1.96 and greater than 1.96.
-  Our observed
value of z falls outside the rejection region, so we fail to reject H0
and conclude that our
drug did not have a significant effect on IQ.

- The IQ is not significantly different than the population mean IQ of 100.
